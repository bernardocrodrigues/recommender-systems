```python
from evaluation import threads
from dataset.movie_lens import resolve_folds
from dataset.movie_lens import load_ml_100k_folds, load_ml_1m_folds
from recommenders import grecond_recommender
from evaluation.threads import grecond_biaknn_benchmark_thread
from evaluation import plot

data, k_fold = load_ml_1m_folds()
folds = resolve_folds(data, k_fold)

parameters = {
    'grecond_coverage': 1.0,
    'dataset_binarization_threshold': 3.0,
    'minimum_bicluster_sparsity':0.001,
    'minimum_bicluster_coverage': None,
    'minimum_bicluster_relative_size': 1e-07,
    'user_binarization_threshold': 3.0,
    'number_of_top_k_biclusters': None,
    'knn_k': 100,
    'knn_type': 'item'
}

recommender_variations = [
    threads.RecommenderVariation(
        "aaa",
        grecond_recommender.GreConDBiAKNNRecommender(
            grecond_coverage=parameters["grecond_coverage"],
            dataset_binarization_threshold=parameters["dataset_binarization_threshold"],
            minimum_bicluster_sparsity=parameters["minimum_bicluster_sparsity"],
            minimum_bicluster_coverage=parameters["minimum_bicluster_coverage"],
            minimum_bicluster_relative_size=parameters["minimum_bicluster_relative_size"],
            knn_type=parameters["knn_type"],
            user_binarization_threshold=parameters["user_binarization_threshold"],
            number_of_top_k_biclusters=parameters["number_of_top_k_biclusters"],
            knn_k=parameters["knn_k"],
        ),
    )
]

results = plot.benchmark(
    folds=folds,
    parallel_recommender_variations=recommender_variations,
    sequential_recommender_variations=[],
    repeats=1,
    relevance_threshold=4.0,
    number_of_top_recommendations=20,
    benchmark_thread=grecond_biaknn_benchmark_thread,
)

# Plot results
from evaluation.threads import GRECOND_BIAKNN_METIC_NAMES

for metric_name in [
    "rmse",
    "micro_averaged_recall",
    "recall_at_k",
    "micro_averaged_precision",
    "precision_at_k",
    "fit_time",
    "test_time",
    "impossible_predictions",
]:
    print(metric_name, plot.get_result_table(metric_name, results))
```
